```python
# ---------------------------------
# 1. Import Libraries
# ---------------------------------
import pandas as pd
from scipy.stats import pearsonr

# ---------------------------------
# 2. Load Cleaned Data
# ---------------------------------
df = pd.read_csv('../data/processed/cleaned_data.csv')

print("Dataset Shape:", df.shape)
df.head()
```

```python
# ---------------------------------
# 3. Correlation: Rating vs Sales
# ---------------------------------
corr_rating_sales, _ = pearsonr(
    df['product_rating'],
    df['purchased_last_month']
)

print("Correlation between Rating and Sales:", corr_rating_sales)
```

```python
# ---------------------------------
# 4. Correlation: Discount vs Sales
# ---------------------------------
corr_discount_sales, _ = pearsonr(
    df['discount_percentage'],
    df['purchased_last_month']
)

print("Correlation between Discount % and Sales:", corr_discount_sales)
```

```python
# ---------------------------------
# 5. Correlation: Reviews vs Sales
# ---------------------------------
corr_reviews_sales, _ = pearsonr(
    df['total_reviews'],
    df['purchased_last_month']
)

print("Correlation between Reviews and Sales:", corr_reviews_sales)
```

```python
# ---------------------------------
# 6. Avg Revenue by Best Seller
# ---------------------------------
best_seller_revenue = df.groupby('is_best_seller')['estimated_revenue'].mean()

print("Average Revenue by Best Seller:\n")
print(best_seller_revenue)
```

```python
# ---------------------------------
# 7. Avg Revenue by Coupon Usage
# ---------------------------------
coupon_revenue = df.groupby('has_coupon')['estimated_revenue'].mean()

print("Average Revenue by Coupon:\n")
print(coupon_revenue)
```

```python
# ---------------------------------
# 8. Avg Revenue by Sponsored Products
# ---------------------------------
sponsored_revenue = df.groupby('is_sponsored')['estimated_revenue'].mean()

print("Average Revenue by Sponsored Products:\n")
print(sponsored_revenue)
```

```python
# ---------------------------------
# 9. Revenue by Rating Category
# ---------------------------------
rating_revenue = df.groupby('rating_category')['estimated_revenue'].mean()

print("Average Revenue by Rating Category:\n")
print(rating_revenue)
```

```python
# ---------------------------------
# 10. Summary Insights (Print)
# ---------------------------------

print("\n--- Key Statistical Observations ---\n")

print("1. Rating vs Sales Correlation:", corr_rating_sales)
print("2. Discount vs Sales Correlation:", corr_discount_sales)
print("3. Reviews vs Sales Correlation:", corr_reviews_sales)

print("\nHigher correlation value means stronger relationship.")
```
