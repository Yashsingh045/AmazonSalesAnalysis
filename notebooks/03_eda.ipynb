```python
# ---------------------------------
# 1. Import Libraries
# ---------------------------------
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ---------------------------------
# 2. Load Cleaned Data
# ---------------------------------
df = pd.read_csv('../data/processed/cleaned_data.csv')

print("Dataset Shape:", df.shape)
df.head()
```

```python
# ---------------------------------
# 3. Basic Information
# ---------------------------------
df.info()
df.describe()
```

```python
# ---------------------------------
# 4. Products by Category
# ---------------------------------
df['product_category'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title("Number of Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------
# 5. Revenue by Category
# ---------------------------------
df.groupby('product_category')['estimated_revenue'].sum().sort_values(ascending=False).plot(kind='bar', figsize=(10,5))
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Estimated Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------
# 6. Rating Distribution
# ---------------------------------
df['product_rating'].hist(bins=10, figsize=(6,4))
plt.title("Product Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# ---------------------------------
# 7. Price Distribution
# ---------------------------------
df['discounted_price'].hist(bins=20, figsize=(6,4))
plt.title("Discounted Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# ---------------------------------
# 8. Reviews vs Sales
# ---------------------------------
plt.figure(figsize=(6,4))
plt.scatter(df['total_reviews'], df['purchased_last_month'])
plt.xlabel("Total Reviews")
plt.ylabel("Sales (Last Month)")
plt.title("Reviews vs Sales")
plt.show()
```

```python
# ---------------------------------
# 9. Rating vs Sales
# ---------------------------------
plt.figure(figsize=(6,4))
plt.scatter(df['product_rating'], df['purchased_last_month'])
plt.xlabel("Product Rating")
plt.ylabel("Sales")
plt.title("Rating vs Sales")
plt.show()
```

```python
# ---------------------------------
# 10. Discount vs Revenue
# ---------------------------------
plt.figure(figsize=(6,4))
plt.scatter(df['discount_percentage'], df['estimated_revenue'])
plt.xlabel("Discount Percentage")
plt.ylabel("Estimated Revenue")
plt.title("Discount vs Revenue")
plt.show()
```

```python
# ---------------------------------
# 11. Best Seller Analysis
# ---------------------------------
df.groupby('is_best_seller')['estimated_revenue'].mean().plot(kind='bar')
plt.title("Average Revenue: Best Seller vs Others")
plt.xlabel("Is Best Seller")
plt.ylabel("Average Revenue")
plt.show()
```

```python
# ---------------------------------
# 12. Correlation Matrix
# ---------------------------------
corr = df[[
    'discounted_price',
    'product_rating',
    'total_reviews',
    'purchased_last_month',
    'discount_percentage',
    'estimated_revenue'
]].corr()

print("Correlation Matrix:\n")
print(corr)
```
