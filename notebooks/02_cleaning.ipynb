```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw/amazon_products_sales_data_raw.csv')
```

```python
print("Initial Shape:", df.shape)
df.head()
```

```python
# -------------------------------
# 3. Basic Exploration
# -------------------------------
df.info()
df.isnull().sum()
```

```python
# -------------------------------
# 4. Remove Duplicates
# -------------------------------
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
# -------------------------------
# 5. Convert Data Types
# -------------------------------

numeric_cols = [
    'discounted_price',
    'original_price',
    'product_rating',
    'total_reviews',
    'purchased_last_month',
    'discount_percentage'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert dates
df['delivery_date'] = pd.to_datetime(df['delivery_date'], errors='coerce')
df['data_collected_at'] = pd.to_datetime(df['data_collected_at'], errors='coerce')
```

```python
# -------------------------------
# 6. Handle Missing Values
# -------------------------------

# Fill specific column
df['buy_box_availability'] = df['buy_box_availability'].fillna('No')

# Drop rows where important values are missing
df = df.dropna(subset=[
    'discounted_price',
    'original_price',
    'product_rating',
    'purchased_last_month'
])

print("After handling missing values:", df.shape)
```

```python
# -------------------------------
# 7. Data Validation & Cleaning
# -------------------------------

# Remove invalid prices
df = df[df['discounted_price'] > 0]
df = df[df['original_price'] > 0]

# Valid rating range (0–5)
df = df[(df['product_rating'] >= 0) & (df['product_rating'] <= 5)]

# Remove extreme discount values (optional but good)
df = df[df['discount_percentage'] <= 90]

print("After validation:", df.shape)
```

```python
# -------------------------------
# 8. Clean Text Columns
# -------------------------------

df['product_category'] = df['product_category'].astype(str).str.strip().str.lower()
```

```python
# -------------------------------
# 9. Standardize Boolean Columns
# -------------------------------

bool_cols = ['is_best_seller', 'is_sponsored', 'has_coupon']

for col in bool_cols:
    df[col] = df[col].astype(str).str.lower()
```

```python
# -------------------------------
# 10. Feature Engineering
# -------------------------------

# Estimated Revenue
df['estimated_revenue'] = df['purchased_last_month'] * df['discounted_price']

# Discount Value
df['discount_value'] = df['original_price'] - df['discounted_price']

# Rating Category
df['rating_category'] = np.where(df['product_rating'] >= 4, 'High', 'Low')
```

```python
# -------------------------------
# 11. Drop Unnecessary Columns
# -------------------------------

df = df.drop(columns=[
    'product_image_url',
    'product_page_url'
], errors='ignore')
```

```python
# -------------------------------
# 12. Final Checks
# -------------------------------

print("Final Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
```
