```python
# ---------------------------------
# 1. Import Libraries
# ---------------------------------
import pandas as pd

# ---------------------------------
# 2. Load Cleaned Data
# ---------------------------------
df = pd.read_csv('../data/processed/cleaned_data.csv')

print("Original Shape:", df.shape)
df.head()
```

```python
# ---------------------------------
# 3. Select Relevant Columns
# ---------------------------------

final_df = df[[
    # Core Analysis
    'product_category',
    'discounted_price',
    'original_price',
    'discount_percentage',
    'product_rating',
    'total_reviews',
    'purchased_last_month',

    # Engineered
    'estimated_revenue',
    'discount_value',
    'rating_category',

    # Business Insight Columns
    'is_best_seller',
    'is_sponsored',
    'has_coupon',
    'buy_box_availability'
]]

print("After Column Selection:", final_df.shape)
```

```python
# ---------------------------------
# 4. Final Checks
# ---------------------------------

print("\nMissing Values:\n")
print(final_df.isnull().sum())

print("\nData Types:\n")
print(final_df.dtypes)
```

```python
# ---------------------------------
# 5. Save Final Dataset
# ---------------------------------

final_df.to_csv('../data/processed/final_tableau_data.csv', index=False)

print("Final Tableau dataset saved successfully!")
```
